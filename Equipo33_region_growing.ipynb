```python

import cv2
import numpy as np
import matplotlib.pyplot as plt


def region_growing(imagen):
  # Cargar la imagen en escala de grises
  image = cv2.imread(imagen, cv2.IMREAD_GRAYSCALE)

  # Copiar la imagen original para visualizarla después
  original_image = image.copy()

  # Crear una máscara con tamaño 2 píxeles más grande en cada dimensión
  height, width = image.shape
  mask = np.zeros((height + 2, width + 2), np.uint8)  # Se agrega +2 en altura y ancho

  # Definir la semilla inicial (ejemplo: punto en el centro de la imagen).
  seed = (width // 2, height // 2) # Se puede definir manualmente con base en experiencia.

  # Para el píxel más oscuro
  #(y, x) = np.unravel_index(np.argmin(image), image.shape)
  #seed = (x, y)

  # Parámetros del crecimiento de la región
  lo_diff = 5  # Diferencia mínima de intensidad permitida
  up_diff = 50  # Diferencia máxima de intensidad permitida
  new_val = 100  # Valor con el que se llenará la región

  # Aplicar el algoritmo de región crecimiento con cv2.floodFill
  cv2.floodFill(image, mask, seedPoint=seed, newVal=new_val, loDiff=(lo_diff,), upDiff=(up_diff,))

  # Mostrar las imágenes
  plt.figure(figsize=(10, 5))

  # Imagen original
  plt.subplot(1, 2, 1)
  plt.imshow(original_image, cmap='gray')
  plt.title('Imagen Original')
  plt.axis('off')

  # Imagen procesada
  plt.subplot(1, 2, 2)
  plt.imshow(image, cmap='gray')
  plt.title('Imagen Procesada')
  plt.axis('off')

  # Mostrar ambas imágenes
  plt.show()

region_growing("toys.png")
```
